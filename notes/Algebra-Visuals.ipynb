# Visualization of Algebra Properties

Lets try to get some hints about Lineal Algebra properties by using $\mathbb{R}^{3 \times 3}$


Lets have the following equation:

$a x_1 + b x_2 + c x_3 = d$

The set of vector [$x_1, x_2, x_3$] such  $a x_1 + b x_2 + c x_3 = d$ is named the solution space.

The solution space of a single equation has one dimention less than the vector space. In our case it is a plane $\mathbb{R}^{2 \times 2}$ in our original $\mathbb{R}^{3 \times 3}$ space.

Lets see the plane form by all the [$x_1, x_2, x_3$] vectors that are solution for: $ x_1 +  x_2 +  x_3 = 0$

```python
from utils import plot_equation, add_axis
ax = plot_equation(1,1,1,0)
add_axis(ax)
```

As you can see this is a plane that intersects [0,0,0], that is because the equation is $=0$.

Any plane intersecting **0** vector is a subspace.

Now if we have two equations:

$$a_1 x_1 + b_1 x_2 + c_1 x_3 = d$$

$$a_2 x_1 + b_2 x_2 + c_2 x_3 = d$$

Means that the solution space are the vectors [$x_1, x_2, x_3$], that makes the two equations true.

In a visual explanation this means that the solution space, is the intersection of two lanes, which is a line:

Let see the solution space for:


$$ x_1 + x_2 + x_3 = 0$$
$$ 10 x_1  + x_3 = 0$$

```python
from utils import plot_equation, add_axis, add_plot_equation
ax = plot_equation(1,1,1,0)
add_plot_equation(ax, 10,0,1,0)
add_axis(ax)
```

The line intersection this two planes is the solution space, as it contains the **0**, then it is a subspace of dimension 1.
