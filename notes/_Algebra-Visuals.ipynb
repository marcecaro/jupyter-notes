# Visualization of Algebra Properties

Lets try to get some graphic hints about Lineal Algebra. The most intuitive way understand it is by using the vector space $\mathbb{R}^{3 \times 3}$. 

Most of the concept that we are going to show in $\mathbb{R}^{3 \times 3}$ can be extrapolated to other dimentions.


## System of Equations

### Solution Space of one equation
Lets have the following equation:

$a x_1 + b x_2 + c x_3 = d$

The set of vector [$x_1, x_2, x_3$] such  $a x_1 + b x_2 + c x_3 = d$ is named the solution space.

The solution space for this equation:

$$x_1 + x_2 + x_3 = 5$$

Is this plane:

```python
from utils import plot_equation, add_axis
ax = plot_equation(1,1,1,5)
add_axis(ax)
```

When $d=0$ the solution space is the plane that intersect the point [$0, 0, 0$].

The solution space for this equation:

$$x_1 + x_2 + x_3 = 0$$

Is the plane:

```python
from utils import plot_equation, add_axis
ax = plot_equation(1,1,1,0)
add_axis(ax)
```


> IMPORTANT: Solution Spaces that includes the point **0**=[$0, 0, 0$] are **Vector Spaces** too.

The solution space of a single equation has one dimention less than the host vector space. 

In our case it is a plane $\mathbb{R}^{2 \times 2}$ in our original $\mathbb{R}^{3 \times 3}$ space.


### Solution Space of two equations


Concider this two equations:

$$a_1 x_1 + b_1 x_2 + c_1 x_3 = d_1$$

$$a_2 x_1 + b_2 x_2 + c_2 x_3 = d_2$$

The solution space are the vectors [$x_1, x_2, x_3$], that makes the two equations true.
In this case is the intersection of two planes, which is a line. 
Moreover if $d_1=0, d_2=0$ the line pass thru **0** which means that it is a **vector space**.

For example:

$$ x_1 + x_2 + x_3 = 0$$
$$ 10 x_1  + x_3 = 0$$

The solution Space is the line resulting from the intersection of these  two planes:

```python
from utils import plot_equation, add_axis, add_plot_equation
ax = plot_equation(1,1,1,0)
add_plot_equation(ax, 10,0,1,0)
add_axis(ax)
```

### Solution Space of three equations


Concider this three equations:

$$a_1 x_1 + b_1 x_2 + c_1 x_3 = d_1$$
$$a_2 x_1 + b_2 x_2 + c_2 x_3 = d_2$$
$$a_3 x_1 + b_3 x_2 + c_3 x_3 = d_3$$

The solution space are the vectors [$x_1, x_2, x_3$], that makes the three equations true.
In this case is the intersection of three planes, which is a point. 
Moreover if $d_1=0, d_2=0, d_3=0$ the point is **0** which means that it is a **vector space**.

For example:

$$ x_1 + x_2 + x_3 = 0$$
$$ 10 x_1  + x_3 = 0$$
$$ 10 x_1  + x_3 = 0$$


The solution Space is the line resulting from the intersection of these  two planes:

```python
from utils import plot_equation, add_axis, add_plot_equation
ax = plot_equation(1,0,1,0)
add_plot_equation(ax, 1,0,0.1,0)
add_plot_equation(ax, 10,1,2,0)
add_axis(ax)
```
